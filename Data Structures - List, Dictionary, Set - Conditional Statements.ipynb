```python
Student_marks = {
    "Venkatesh": 75,
    "Radha": 68,
    "Anitha": 90,
    "Sheela": 85,
    "Chandu": 72
}
print(Student_marks)
```

```python
print(Student_marks["Sheela"])
```

```python
Student_marks["Janani"] = 80
print(Student_marks)
```

```python
Student_marks["Chandu"] = 82
print(Student_marks)
```

```python
print("Keys:", Student_marks.keys())
print("Values:", Student_marks.values())
print("Items:", Student_marks.items())
```

```python
my_set = {'a','e','i','o','u','a','a','i'}
print(my_set)
```

```python
# Set is a collection of unique values.It automatically removes duplicates.In the given list 'a' and 'i' are duplicate values so itis removed.
# since set is unordered output order may change
```

```python
my_set[4] = 's'
```

```python
# Sets are unordered,python do not support indexing, we cannot access elements using position like [0], [1]
```

```python
set1 = {1,3,5,7,9}
set2 = {2,3,5,8,10}

print(set1)
print(set2)
```

```python
print(set1.union(set2))
print(set1.intersection(set2))
```

```python
score = int(input("Enter your score (0 to 10): "))

if score < 0 or score > 10:
    print("Invalid score. Please enter value between 0 and 10.")
elif score > 7:
    print("Above Average: Excellent work! Keep it up.")
elif score >= 4:
    print("Average: Good effort! Keep practicing.")
else:
    print("Below Average: Need to improve your performance.")
```
