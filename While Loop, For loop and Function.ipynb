```python
# 1.Create a Python program that implements a simple number
#guessing game using a while loop. The program should make use of control
#statements such as else, break, and continue.

import random

secret_number = random.randint(1, 10)

while True:
    guess = int(input("Enter a number between 1 and 10: "))

    if guess < 1 or guess > 10:
        print("Invalid input! Try again.")
        continue

    if guess != secret_number:
        print("Wrong guess!")
        continue

    print("Correct guess!")
    break
```

```python
# 2.Prompt the user using variable attempts
import random

secret_number = random.randint(1, 10)
attempts = 3

while attempts > 0:
    guess = int(input("Guess the number (1-10): "))

    if guess == secret_number:
        print("Correct! You guessed it.")
        break
    else:
        attempts = attempts - 1
        print("Wrong guess! Attempts left:", attempts)

else:
    print("Game Over! The number was:", secret_number)
```

```python
# 3.Implement the Guessing Logic(while)
import random

secret_number = random.randint(1, 10)
attempts = 3

while attempts > 0:
    guess = int(input("Guess a number (1-10): "))
    if guess < 1 or guess > 10:
        print("Invalid input! Please enter between 1 and 10.")
        continue
    if guess == secret_number:
        print("Congratulations! You guessed the correct number.")
        break
    elif guess > secret_number:
        print("Too high!")
    else:
        print("Too low!")
    attempts = attempts - 1
    print("Attempts left:", attempts)
else:
    print("Game Over! The number was:", secret_number)
```

```python
# 4.Control Statements(continue, break, else, while)
import random

secret_number = random.randint(1, 10)
attempts = 3

while attempts > 0:
    guess = int(input("Guess a number (1-10): "))

    if guess < 1 or guess > 10:
        print("Invalid input! Please enter between 1 and 10.")
        continue
    if guess == secret_number:
        print("Congratulations! You guessed the correct number.")
        break
    elif guess > secret_number:
        print("Too high!")
    else:
        print("Too low!")

    attempts = attempts - 1
    print("Attempts left:", attempts)

else:
    print("Better luck next time! The number was:", secret_number)
```

```python
# Create a Python program that generates and prints a
#multiplication table (from 1 to 10) for a given number using a for loop and the
#range function using for loop

Num = int(input("Enter the number for which you want the multiplication table: "))

for i in range(1, 11):
    result = Num * i
    print(Num, "x", i, "=", result)
```

```python
#Create a Python program that calculates the Body Mass Index (BMI) using function

def calculate_bmi(weight, height):
    return weight / (height ** 2)

weight = float(input("Enter your weight in kg: "))
height = float(input("Enter your height in meters: "))

bmi = calculate_bmi(weight, height)

print("Your BMI is:", round(bmi, 2))
```
