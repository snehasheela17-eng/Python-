Pandas Series:
1. Creating Pandas Series:
 Create a Pandas Series named marks with the following data

```python
import pandas as pd

marks = pd.Series(
    [95, 92, 89, 85, 80],
    index=['Rank1', 'Rank2', 'Rank3', 'Rank4', 'Rank5']
)

print(marks)
```

2. Indexing and Slicing:
Use the integer index position to access the mark of the 1st rank student.

```python
import pandas as pd

marks = pd.Series(
    [95, 92, 89, 85, 80],
    index=['Rank1', 'Rank2', 'Rank3', 'Rank4', 'Rank5']
)

print("1st rank:", marks[0])
```

Use the loc accessor to retrieve the marks of the top 3 ranks by
specifying their index labels.

```python
print("Top 3 ranks:")
print(marks.loc['Rank1':'Rank3'])
```

Use the iloc accessor to retrieve the mark of the 3rd rank student.

```python
print("3rd rank mark:", marks.iloc[2])
```

Apply a boolean mask to filter the Series and retrieve the ranks where the
marks are greater than 90.

```python
print("Marks greater than 90:")
print(marks[marks > 90])
```

3. Manipulating Series:

```python
marks.loc['Rank1'] = 100
print("After updating Rank1:")
print(marks)
```

Remove the entry corresponding to the last rank from the Series.


```python
marks = marks.drop('Rank5')
print("After removing last rank:")
print(marks)
```

Compute the CGPA by dividing each mark in the Series by 10.

```python
cgpa = marks / 10
print("CGPA:")
print(cgpa)
```

Pandas DataFrame:
1. Creating Pandas DataFrame:
Create a Pandas DataFrame named transactions with the following data:

```python
import pandas as pd

transactions = pd.DataFrame({
    'TransactionID': [101, 102, 103, 104, 105, 106, 107, 108, 109, 110]
})

print(transactions)
```

ProductCategory: 'Electronics', 'Clothing', 'Electronics', 'Furniture',
'Clothing', 'Electronics', 'Furniture', 'Clothing', 'Furniture', 'Electronics'

```python
transactions['ProductCategory'] = [
    'Electronics', 'Clothing', 'Electronics', 'Furniture',
    'Clothing', 'Electronics', 'Furniture', 'Clothing',
    'Furniture', 'Electronics'
]

print(transactions)
```

Region: 'North', 'South', 'North', 'East', 'West', 'North', 'East', 'West',
'South', 'North'

```python
region = [
    'North', 'South', 'North', 'East', 'West',
    'North', 'East', 'West', 'South', 'North'
]

transactions['Region'] = region
print(transactions)
```

Amount: 200, 150, 300, 450, 200, 250, 300, 180, 350, 400

```python
transactions['Amount'] = [200, 150, 300, 450, 200, 250, 300, 180, 350, 400]

print(transactions)
```

2. Data Exploration:
Display the transactions DataFrame and its basic information, including
head, tail, shape, column names, and data types.

```python
print("Complete DataFrame:")
print(transactions)

print("\nHead:",transactions.head())

print("\nTail:",transactions.tail())

print("\nShape:",transactions.shape)

print("\nColumn Names:",transactions.columns)

print("\nData Types:",transactions.dtypes)
```

Display only the 'ProductCategory' and 'Amount' columns.


```python
print(transactions[['ProductCategory', 'Amount']])
```

Retrieve the last 3 columns of the DataFrame using iloc or loc.


```python
print(transactions.iloc[:, 1:])
```

Filter rows where the 'Region' is 'North' and 'Amount' is greater than 200.


```python
filtered_rows = transactions[
    (transactions['Region'] == 'North') &
    (transactions['Amount'] > 200)
]

print(filtered_rows)
```

Find the value counts for the 'ProductCategory' column.


```python
print(transactions['ProductCategory'].value_counts())
```

Find the unique values in the 'Region' column.


```python
print(transactions['Region'].unique())
```

Group by 'Region' and find the mean amount for each region.


```python
print(transactions.groupby('Region')['Amount'].mean())
```

3. Manipulating the DataFrame:
Modify the 'Amount' for the transaction with TransactionID 102 to 165.

```python
transactions.loc[transactions['TransactionID'] == 102, 'Amount'] = 165

print(transactions)
```

Add a new column 'Discount' by calculating 10% of the 'Amount'


```python
transactions['Discount'] = transactions['Amount'] / 10
print(transactions)
```

Remove the row with TransactionID 109.

```python
transactions = transactions.drop(
    transactions[transactions['TransactionID'] == 109].index
)
print(transactions)
```

Delete the 'Discount' column from the DataFrame.

```python
del transactions['Discount']

print(transactions)
```
